This script will intake a text file of energies and RMS from AMBER minimization

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This function reads the data from a file and returns the step, energy, and rms arrays
def plot_energy(step, energy, input_file):
    """Plots step vs energy and uses the file name as the title."""
    file_name = os.path.basename(input_file)  # Extract the file name from the path
    plt.figure(figsize=(5, 5))
    plt.plot(step, energy, marker='o', color='b', label='Energy')
    plt.xlabel('Step')
    plt.ylabel('Energy')
    plt.title(f'Step vs Energy - TrwD_atp')
    plt.grid(True)
    plt.show()

def plot_rms(step, rms, input_file):
    """Plots step vs RMS and uses the file name as the title."""
    file_name = os.path.basename(input_file)  # Extract the file name from the path
    plt.figure(figsize=(5, 5))
    plt.plot(step, rms, marker='o', color='r', label='RMS')
    plt.xlabel('Step')
    plt.ylabel('RMS')
    plt.title(f'Step vs RMS - TrwD_atp')
    plt.grid(True)
    plt.show()

def load_data(input_file):
    """Loads data from a file and returns step, energy, and rms arrays."""
    data = np.genfromtxt(input_file, delimiter=',', skip_header=1)
    step = data[:, 0]
    energy = data[:, 1]
    rms = data[:, 2]
    return step, energy, rms

def plot_rms_shorter(step, rms, input_file):
    file_name = os.path.basename(input_file)  # Extract the file name from the path
    plt.figure(figsize=(5, 5))
    plt.plot(step, rms, marker='o', color='r', label='RMS')
    plt.xlabel('Step')
    plt.xlim(5000)  # X-axis limit stays the same
    plt.ylabel('RMS')
    plt.title(f'Step vs RMS - TrwD_atp')
    plt.grid(True)

    # Check the range of RMS values
    min_rms = min(rms)
    max_rms = max(rms)

    # Set the y-axis limit with the lower limit set to the min RMS value and the upper limit to 1000
    plt.ylim(min_rms-0.01, 0.4)  # Adjust the range as necessary

    plt.show()
```

```python
input_file = '/Users/stuytschaevers/Desktop/Thesis/atp_mg2+/cluster/min_energies/TrwD_atp_energy_rms_values.csv'

step, energy, rms = load_data(input_file)
plot_energy(step, energy, input_file)
plot_rms(step, rms, input_file)
plot_rms_shorter(step, rms, input_file)
```

```python
# Function to plot step vs potential energy
def plot_energy(step, energy, input_file, title):
    """Plots step vs energy and uses the file name as the title."""
    plt.figure(figsize=(5, 5))
    plt.plot(step, energy, marker='o', color='b', label='Potential Energy')
    plt.xlabel('Step')
    plt.ylabel('Potential Energy (kJ/mole)')
    plt.title(f'Step vs Potential Energy - {title}')
    plt.grid(True)
    plt.show()

# Function to load data (Step, Potential Energy, Temperature)
def load_data(input_file):
    """Loads data from a file and returns step and energy arrays."""
    data = np.genfromtxt(input_file, delimiter=',', skip_header=1)
    step = data[:, 0]
    energy = data[:, 1]
    return step, energy

# Example usage for plotting energy vs step
def main(input_file, title):
    step, energy = load_data(input_file)
    plot_energy(step, energy, input_file, title)

# Example input file (you should provide the actual file path)
pose50 = "/Users/stuytschaevers/Desktop/Thesis/atp_mg2+/amber/test/from_cluster/output/min_energies/pose_50_energies.csv"
pose241 = "/Users/stuytschaevers/Desktop/Thesis/atp_mg2+/amber/test/from_cluster/output/min_energies/pose_241_energies.csv"

main(pose50, "Pose 50")
main(pose241, "Pose 241")
```
